\begin{align*}
\mathcal{F}\big[ax[n] + by[n] \big](\omega) &= \sum_{n=-\infty}^{\infty}\big[ax[n] + by[n] \big]e^{-i\omega n}\\
 &= \; a\sum_{n=-\infty}^{\infty}x[n]e^{-i\omega n}+ b\sum_{n=-\infty}^{\infty}y[n]e^{-i\omega n}\\
 &=  \;a\,X(\omega) + b\,Y(\omega)\\
 &=  \;a\,\mathcal{F}\big[x[n]\big](\omega) + b\,\mathcal{F}\big[y[n]\big](\omega)
\end{align*}

Nechť vzor $f: \mathbb{R} \rightarrow \mathbb{C}$ je Lebesguovsky integrovatelná funkce na $(-\infty, \infty)$, potom Fourierovým obrazem $f$ je funkce:  

$$F(i\omega) = \mathcal{F}\left[f(t)\right](i\omega) :=  \int_{-\infty}^{+\infty} f(t) e^{-i\omega t} \text{d}t, \omega \in \mathbb{R}$$

```python
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(25,5))
ax1 = fig.add_subplot(121); ax2 = fig.add_subplot(122)

o = np.linspace(-2*np.pi, 2*np.pi, 200)
X = np.cos(.4*np.pi*o) + np.cos(.5*np.pi*o)**2 - .2*o**2+7.12
ax1.plot(o,X, color='blue')
ax1.set_xlabel('$\omega$')
ax1.set_ylabel('$X(omega)$')


X = np.cos(.4*np.pi*o*3) + np.cos(.5*np.pi*o*3)**2 - .2*(3*o)**2 + 7.12
X2 = np.cos(.4*np.pi*(o+4*np.pi/3)*3) + np.cos(.5*np.pi*(o+4*np.pi/3)*3)**2 - .2*(3*(o+4*np.pi/3))**2 + 7.12
X1 = np.cos(.4*np.pi*(o-4*np.pi/3)*3) + np.cos(.5*np.pi*(o-4*np.pi/3)*3)**2 - .2*(3*(o-4*np.pi/3))**2 + 7.12
ax2.plot(o,X, color='green')
ax2.plot(o,X1, color='green')
ax2.plot(o,X2, color='green')
ax2.set_ylim([0,10.5])
ax2.set_xlabel('$\omega$')
ax2.set_ylabel('$X(l\omega)$')
plt.show()
```
